```python

# نصب پکیج‌های لازم
!pip install spleeter
!pip install pydub
!apt install ffmpeg -y

# ایمپورت کتابخانه‌ها
from spleeter.separator import Separator
from pydub import AudioSegment
import os

# مسیر فایل آهنگ در گوگل درایو
input_audio_path = '/content/drive/MyDrive/song.mp3'

# مسیر خروجی
output_path = '/content/drive/MyDrive/spleeter_output'
song_name = 'song'

# ساخت پوشه خروجی
os.makedirs(output_path, exist_ok=True)

# جداسازی صدا و موسیقی
separator = Separator('spleeter:2stems')
separator.separate_to_file(input_audio_path, output_path)

# مسیر فایل‌های WAV که توسط Spleeter ساخته می‌شن
vocals_wav = f'{output_path}/{song_name}/vocals.wav'
music_wav = f'{output_path}/{song_name}/accompaniment.wav'

# مسیر نهایی برای ذخیره MP3
vocals_mp3 = f'{output_path}/{song_name}/vocals.mp3'
music_mp3 = f'{output_path}/{song_name}/accompaniment.mp3'

# تبدیل WAV به MP3 با استفاده از pydub
AudioSegment.from_wav(vocals_wav).export(vocals_mp3, format="mp3")
AudioSegment.from_wav(music_wav).export(music_mp3, format="mp3")

print("✅ صدای خواننده و موسیقی با موفقیت به فرمت MP3 ذخیره شدند!")
print("📁 مسیر فایل خواننده:", vocals_mp3)
print("🎵 مسیر فایل موسیقی:", music_mp3)
```
